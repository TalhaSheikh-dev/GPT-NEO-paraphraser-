```python
def prompt_selector(input_text,level):
  prompt_text = ""
  total_length = 0
  input_length =0
  if level == "1":
    prompt_text = para1.format(input_text)
    if len(input_text) <70:
      total_length = len(input_text)*0.7
    elif len(input_text) <100:
      total_length = len(input_text)*0.6
    elif len(input_text) <150:
      total_length = len(input_text)*0.5
    else:
      total_length = len(input_text)*0.55
    total_length = total_length+len(prompt_text.split(" "))*1.35
    input_length = len(prompt_text.split(" "))*1.35

  elif level == "2":
    prompt_text = impro.format(input_text)
    if len(input_text) <70:
      total_length = len(input_text)*0.8
    elif len(input_text) <100:
      total_length = len(input_text)*0.7
    elif len(input_text) <150:
      total_length = len(input_text)*0.6
    else:
      total_length = len(input_text)*0.55
    total_length = total_length+len(prompt_text.split(" "))*1.3
    input_length = len(prompt_text.split(" "))*1.3
  return prompt_text,total_length,input_length
```

```python
def clean(text,level):
  text1=""
  if level =="1":
    text1 = text.split("Paraphrase:")[-1]
  elif level == "2":
    text1 = text.split("Intro:")[-1]
  m = re.finditer(r'\.\s[A-Za-z]', text1)
  hi = 0
  check = 0
  for a in m:
    check=1
    hi = max(hi,a.end(0))
  if check == 1:
    text1 = text1[:hi-2]
  else:
    pass
  
  
  return text1.strip()
```

```python
def read_data(name):
  df = pd.read_excel(name)
  df = df[["input","output"]]
  df = df[df['input'].apply(lambda x: len(str(x)) >= 40)]
  return df
```

```python
def character_length(text):
  try:
    all = text.split(" ")
    temp=""
    for a in all:
      if len(a)>15:
        temp = re.sub(",","",a)
        temp = re.sub("\.","",temp)
        temp = re.sub(";","",temp)
        temp = re.sub(":","",temp)
        temp = re.sub("\?","",temp)
        temp = re.sub("[0-9]","",temp)
        if temp.isalpha():
          return 1
        if len(a) > 30:
          return 1
  except:
    pass
  return 0
```

```python
def captial_calculator(string):
  total_length = len(string.split(" "))
  count1=0
  for i in string:
        if(i.isupper()):
              count1=count1+1
  if count1>0.5*total_length:
    return 1
  return 0 
```

```python
def get_paraphrase_of_parapghrape(paragraph,Creativity,five_counter):
  text = ""
  paragraph = paragraph.strip(" ").strip("\n")
  prompt_text,output_length,input_length = prompt_selector(paragraph,Creativity)

  text= ai_text(prompt_text,output_length,input_length)
  text = re.sub('\s+',' ',text)
  text = clean(text,Creativity)
  if (character_length(text) or  captial_calculator(text)) and five_counter<5:
    five_counter = five_counter+1
    return get_paraphrase_of_parapghrape(text,Creativity,five_counter) 
  if character_length(text) and five_counter==2:
    return None


  return text
```

```python
def question_remover(text):
  try:
    quest = text.index("?")
    try:
      dot = text.index(".")
    except:
      dot = len(text)

    if quest< dot:
      return text[quest+1:]
  except:
    return text
  return text
```

```python
import re

def capitalize_string(s):
    try:
        m = re.search(r'[a-z]', s)
        m1 = re.search(r'[A-Z]', s)
        if m is not None:
            if m1 is not None:
                if m.start()<m1.start():
                    S = list(s)
                    S[m.start()] = S[m.start()].capitalize()
                    s = ''.join(S)
            else:
                S = list(s)
                S[m.start()] = S[m.start()].capitalize()
                s = ''.join(S)
        if s[-1] != ".":
          s = s +"."
    except:
        return s
    return s
```

```python
def break_text(text,Creativity):
    output_list = re.findall(r'<[^>]*>.*?</[^>]*>(?:<[^>]*/>)?|[^<>]+', text)
    tags = []
    paragraphs = []
    for a  in output_list:
      x = a.strip("\n").strip(" ").strip("\t")
      if x[0] == "<":
        tags.append(x)
      else:
        paragraphs.append(x)
    out_temp = ""
    out_ans = ""
    new_ans = ""
    for i in range(len(paragraphs)):
        te = paragraphs[i]
        te = re.sub("\n","",te)
        if te == "":
            continue
        if i == 0:
            out_ans = get_paraphrase_of_parapghrape(te,Creativity,0)
            out_ans = question_remover(out_ans).strip(";").strip(",").strip(".").strip(" ")
            out_ans = capitalize_string(out_ans)
            out_temp = out_temp + out_ans+"\n\n"
            continue
        tag = tags[i-1]
        new_ans = get_paraphrase_of_parapghrape(te,Creativity,0)
        if new_ans == None or (re.sub(" ","",re.sub("\n","",new_ans)) == ""):
            continue
        out_temp = out_temp + tag+"\n"
        new_ans = question_remover(new_ans).strip(";").strip(",").strip(".").strip(" ")
        new_ans = capitalize_string(new_ans)
        out_temp = out_temp + new_ans+"\n\n"
    try:
        temp_one = out_temp
        a = re.search("<h2>(.|\n^<h2>)*?</h2>\n*<h2>",temp_one)
        temp_one = temp_one[:a.start()] +" "+temp_one[a.end()-4:]
        out_temp = temp_one
    except:
        pass
    return out_temp
```

```python
def complete_process(df,Creativity):
  saving ="/content/output.xlsx"
  try:
    os.remove(saving)
  except:
    pass
  paragraph_income = ""
  saving_df = pd.DataFrame(columns=["input","output"])
  row = ["",""]
  total_df_len = len(df)
  check_counter = 1
  print("starting processing")
  for article in df.iterrows():
    print("the processing is done {}/{}".format(check_counter,total_df_len))
    row = ["",""]
    row[0] = article[1][0]
    paragraph_income = article[1][0]
    ans_answer_final = break_text(paragraph_income,Creativity)
    row[1] = ans_answer_final
    df_length = len(saving_df)
    saving_df.loc[df_length] = row
    saving_df.to_excel(saving)
    check_counter = check_counter+1
    
  return "done with processing"
```
